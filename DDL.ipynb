
# DDL - Data Definition Language

DDL (Data Definition Language) commands are used to define, create, modify, and delete database objects such as:
- Tables
- Databases
- Schemas
- Views

DDL commands mainly work on the structure of database objects rather than the actual data.

---

# Characteristics of DDL Commands

- Used to manage database schema
- Automatically committed in most databases
- Changes are permanent
- Affects metadata
- Mainly used by Data Engineers and DBAs

---

# Common DDL Commands

1. CREATE
2. ALTER
3. DROP
4. TRUNCATE
5. RENAME

---

# 1. CREATE Command

## Definition
`CREATE` command is used to create new database objects.

Objects that can be created:
- Tables
- Databases
- Views
- Schemas

---

## CREATE TABLE Syntax

```sql
CREATE TABLE table_name (
    column1 datatype,
    column2 datatype,
    column3 datatype
);
```

---

## CREATE TABLE Example

```sql
CREATE TABLE employee (
    emp_id INT,
    emp_name STRING,
    salary DOUBLE,
    joining_date DATE
);
```

---

## CREATE TABLE with Constraints

```sql
CREATE TABLE employee (
    emp_id INT PRIMARY KEY,
    emp_name STRING NOT NULL,
    salary DOUBLE,
    department STRING DEFAULT 'IT'
);
```

---

## CREATE DATABASE Example

```sql
CREATE DATABASE company_db;
```

---

## CREATE VIEW Example

```sql
CREATE VIEW employee_view AS
SELECT emp_id, emp_name
FROM employee;
```

---

## CREATE TABLE USING DELTA

```sql
CREATE TABLE sales (
    order_id INT,
    customer_name STRING,
    amount DOUBLE
)
USING DELTA;
```

---

## CREATE TABLE AS SELECT (CTAS)

```sql
CREATE TABLE employee_backup AS
SELECT * FROM employee;
```

---

## Important Points about CREATE

- Creates database objects from scratch
- Defines schema structure
- Supports constraints
- Commonly used in Databricks Delta Lake

---

# 2. ALTER Command

## Definition
`ALTER` command is used to modify the structure of an existing table.

Using ALTER we can:
- Add columns
- Drop columns
- Rename columns
- Change data types
- Rename tables

---

## ADD COLUMN Example

```sql
ALTER TABLE employee
ADD COLUMN department STRING;
```

---

## ADD MULTIPLE COLUMNS Example

```sql
ALTER TABLE employee
ADD COLUMNS (
    location STRING,
    manager_id INT
);
```

---

## DROP COLUMN Example

```sql
ALTER TABLE employee
DROP COLUMN department;
```

---

## RENAME COLUMN Example

```sql
ALTER TABLE employee
RENAME COLUMN emp_name TO employee_name;
```

---

## CHANGE DATA TYPE Example

```sql
ALTER TABLE employee
ALTER COLUMN salary TYPE BIGINT;
```

---

## ALTER TABLE RENAME Example

```sql
ALTER TABLE employee
RENAME TO employee_master;
```

---

## Important Points about ALTER

- Modifies existing schema
- Existing data remains unchanged
- Useful for schema evolution
- Common in real-time projects

---

# 3. DROP Command

## Definition
`DROP` command is used to permanently remove database objects.

Objects that can be dropped:
- Tables
- Databases
- Views
- Schemas

---

## DROP TABLE Syntax

```sql
DROP TABLE table_name;
```

---

## DROP TABLE Example

```sql
DROP TABLE employee;
```

---

## DROP TABLE IF EXISTS Example

```sql
DROP TABLE IF EXISTS employee;
```

---

## DROP DATABASE Example

```sql
DROP DATABASE company_db;
```

---

## Important Points about DROP

- Removes both structure and data
- Permanent operation
- Cannot be recovered easily
- Use carefully in production

---

# 4. TRUNCATE Command

## Definition
`TRUNCATE` command removes all rows from a table but keeps the table structure.

---

## TRUNCATE Syntax

```sql
TRUNCATE TABLE table_name;
```

---

## TRUNCATE Example

```sql
TRUNCATE TABLE employee;
```

---

## Important Points about TRUNCATE

- Deletes all records quickly
- Table structure remains
- Faster than DELETE
- WHERE clause cannot be used

---

## Difference Between DELETE and TRUNCATE

| DELETE | TRUNCATE |
|---|---|
| Removes selected rows | Removes all rows |
| WHERE clause allowed | WHERE clause not allowed |
| Slower | Faster |
| Row-level logging | Minimal logging |
| Structure remains | Structure remains |

---

# 5. RENAME Command

## Definition
Used to rename database objects like tables or columns.

---

## RENAME TABLE Example

```sql
ALTER TABLE employee
RENAME TO employee_details;
```

---

## RENAME COLUMN Example

```sql
ALTER TABLE employee_details
RENAME COLUMN emp_name TO employee_name;
```

---

## Important Points about RENAME

- Changes only object name
- Data remains unchanged
- Helps maintain naming standards

# Constraints in SQL

| Constraint | Purpose |
|---|---|
| PRIMARY KEY | Unique identifier |
| NOT NULL | Prevents null values |
| UNIQUE | Allows unique values |
| DEFAULT | Assigns default value |
| CHECK | Validates condition |
| FOREIGN KEY | Maintains relationships |

# Real-Time Use Cases in Databricks

## CREATE
- Creating Delta tables
- Creating Bronze/Silver/Gold layer tables
- Creating temporary views

## ALTER
- Adding new business columns
- Schema evolution
- Modifying data types

## DROP
- Removing unused staging tables
- Cleaning old backup tables

## TRUNCATE
- Clearing staging tables before reload

## RENAME
- Standardizing naming conventions

---

# Best Practices

- Use meaningful table names
- Use `IF EXISTS` while dropping tables
- Prefer Delta tables in Databricks
- Avoid DROP in production without backup
- Maintain schema consistency

# Summary

DDL commands are important in SQL and Databricks for managing database structures.

Main DDL Commands:
- CREATE → Create database objects
- ALTER → Modify existing objects
- DROP → Remove objects permanently
- TRUNCATE → Remove all data only
- RENAME → Rename database objects

These commands are heavily used in real-time Data Engineering projects for schema management and Delta Lake operations.
